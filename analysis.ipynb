```python
import numpy
import pandas
import json
```

```python
f = open('users20221112.json')
data = json.load(f)
print(data)
```

```python
played = {}
won = {}
for key in data.keys():
    success = data[key]['success']
    for succ in success.keys():
        played[succ] = played.get(succ, 0) + 1
        if(success[succ]):
            won[succ] = won.get(succ, 0) + 1
print(played)
print(won)
        
```
